```python
from sklearn import datasets
import pandas as pd
```

```python
iris = datasets.load_iris()
```

```python
dir(iris)
```

```python
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
```

```python
df['species_id'] = pd.Series(iris.target)
df['species_name'] = pd.Series([iris.target_names[x] for x in iris.target])
```

```python
df
```

```python
%matplotlib inline
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
for l in iris.target_names:
    sns.distplot(df[df['species_name'] == l]['sepal length (cm)'], label=l)
plt.legend()
plt.show()
```

```python
for l in iris.target_names:
    sns.distplot(df[df['species_name'] == l]['sepal width (cm)'], label=l)
plt.legend()
plt.show()
```

```python
for l in iris.target_names:
    sns.distplot(df[df['species_name'] == l]['petal length (cm)'], label=l)
plt.legend()
plt.show()
```

```python
for l in iris.target_names:
    sns.distplot(df[df['species_name'] == l]['petal width (cm)'], label=l)
plt.legend()
plt.show()
```
